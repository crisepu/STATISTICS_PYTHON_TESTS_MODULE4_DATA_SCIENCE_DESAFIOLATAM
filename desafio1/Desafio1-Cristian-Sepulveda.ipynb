**1.** El promedio general de los salarios, desviaci√≥n est√°ndar, quintiles y rango.

```python
promedio_salarios = df_salaries['salary_in_usd'].mean()
desviacion_estandar = df_salaries['salary_in_usd'].std()
quintiles = df_salaries['salary_in_usd'].quantile([0.2, 0.4, 0.6, 0.8])
rango = df_salaries['salary_in_usd'].max() - df_salaries['salary_in_usd'].min()

print(f'Promedio Salarios (USD): {promedio_salarios}')
print(f'Desviacion Estandar (USD): {desviacion_estandar}')
print(f'Quintiles: \n{quintiles}')
print(f'Rango (USD): {rango}')
```

**2.** Una comparaci√≥n entre los salarios agrupados por 3 diferentes categor√≠as a tu elecci√≥n. ¬øPara cu√°l de ellas las medidas de tendencia central son m√°s representativas? ¬øPara cu√°l son menos representativas? Explica.

*2.1* En primer lugar agruparemos el salario en dolares segun el nivel de experiencia:

```python
#Primero haremos un primero arreglando, ajustando la el nivel de experience en orden de mayor a experiencia a menos experiencia.
level_order = ['EN', 'MI', 'SE', 'EX']
df_salaries['experience_level'] = pd.Categorical(df_salaries['experience_level'], categories=level_order, ordered=True)

#Generamos tabla pivote para analisis de medidas de tendencia central en el agrupamiento indicado
tabla_pivote1 = pd.pivot_table(df_salaries, values=['salary_in_usd'], index='experience_level', aggfunc=['mean', 'std', 'min', 'max'])
display(tabla_pivote1)
```

Con la tabla, es posible apreciar que en terminos generales la media es representativa de cada agrupamiento segun el nivel de experiencia, podemos apreciar que a medida aumenta la experiencia, el promedio de los salarios aumenta. LA desviacion estandar es constantes en los niveles de experiencia junior (EN), intermedio (MI) y senior (SE), aumentando considerablemente alrededor de un 22% en el nivel de experiencia experto (EX). Tambien en los rangos se aprecian casos particulaes en el nivel de experiencia intermedio en donde el salario minimo es menor que el del junior y el salario maximo es mayor que el del senior o experto, lo que denota una alta dispersion en este punto.

*2.2* La segunda agrpacion que haremos es el salario en dolares segun cargo

```python
#Generamos tabla pivote para analisis de medidas de tendencia central en el agrupamiento indicado
tabla_pivote2 = pd.pivot_table(df_salaries, values=['salary_in_usd'], index='job_title', aggfunc=['mean', 'std', 'min', 'max'])
display(tabla_pivote2)
```

En este caso las medidas de tendencia central no son representativas ya que la desviacion estandar y los rangos maximos y minimos se alejan mucho de la media, esto probablemente debido a que solo estamos agrupando por cargo, sin considerar los a√±os de experiencia, lo quegenera una alt√≠sima dispersion en cada grupo.

*2.3* Y en tercer lugar agruparemos el salario en dolares segun el ratio de trabajo remoto

```python
#Generamos tabla pivote para analisis de medidas de tendencia central en el agrupamiento indicado
tabla_pivote3 = pd.pivot_table(df_salaries, values=['salary_in_usd'], index='remote_ratio', aggfunc=['mean', 'std', 'min', 'max'])
display(tabla_pivote3)
```

En este caso podemos cer que tanto el grupo sin trabajo remoto con el  totalmente remoto poseen una emdia no tan distinta y una desviacion estandar similar, no asi el parcialmente remoto, que el promedio es menor a los otros dos agrupamientos y la desbiacion estandar es considerable en relacion al promedio. Considero que para este agrupamiento las medidas de tendencia central no son representativas.

**3.** ¬øQu√© cargos reciben mejores sueldos en las empresas con sede en Estados Unidos?

```python
#PAra esto primero generaremos un solo dataframe, el cual filtrara los datos con las empresas con sede en US 
df_salaries_us = df_salaries.query('company_location == "US"')
df_salaries_us
```

```python
#Agrupamos el salario en usd por cargo e imprimimemos el resultado de los 10 mayores salarios en orden descedente
mean_salary_by_job= df_salaries_us.groupby('job_title')['salary_in_usd'].mean()
display(mean_salary_by_job.sort_values(ascending=False).head(10))
```

Como resultado podemos ver que el cargo con mejor sueldo en promedio. considerando solo las empresas con sede en Estdos Unidos, es el de Data Analytics Lead, siguiendo con los roles de Data Science Tech Lead y Director of Data Science, en segundo y tercer lugar respectivamente.
